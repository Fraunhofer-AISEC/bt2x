```python

import os
import matplotlib.pyplot as plt
import json
```

## Libs that are dependencies of both TUF and BT code

The folowing libraries are mostly used by both and therefore cannot be assigned as dependencies for one of the specific features.  


```python
crypto_libs = [
    "p256",
    "sha2",
    "pem_rfc7468",
    "spki",
    "pkcs8",
    "der",
    "der?",
    "x509_cert",
    "serde_json",
    "const_oid",
    "ecdsa",
    "base64ct",
    "sec1",
    "crypto_bigint",
    "heapless",
    "smoltcp",
    "embassy_net",
    "fixed",
    "hash32",
]
```

## Flash Size and WiFi Drivers

- The flash size is listed as 2MB, which we convert to MiB
- The drivers for the WiFi chip on the Pi Pico W requires drivers that need to flashed onto the same flash chip as the binary itself. It can also be "baked" into the binary directly, however this makes it impossible to reuse the wifi drivers that are already stored on the chip for new binaries when updating. It also increases the size of the binaries significantly.

```python
wifi_driver_size = os.stat(f"../-pi-pico-w/firmware/43439A0_clm.bin").st_size + os.stat("../-pi-pico-w/firmware/43439A0.bin").st_size
wifi_driver_size /= 1024 # scale to KiB

flash_size = 2 * 1e6 / 1024**1
```

## Colors

```python
flash_size_data_color = "tab:blue"
wifi_driver_size_data_color = "tab:red"
file_size_data_color = "tab:olive"
file_text_section_color ="tab:pink"
tuf_no_std_size_color = "tab:cyan"
_size_color="tab:orange"
libs_size_color="tab:purple"
```

```python
data_path = "../benchmark-output/benchmarks-size.json"

plt.style.use("Solarize_Light2")
print("processing size benchmarks")
with open(data_path) as f:
    raw_data = json.load(f)


file_size_data = [(name, data["raw_size"]/(1024**1)) for (name, data) in raw_data.items()]
flash_size_data = [(name, flash_size) for (name, _) in raw_data.items()]
file_text_section = [(name, data["cargo_bloat_crates"]["text-section-size"]/(1024**1)) for (name,data) in raw_data.items()]
tuf_no_std_size = []
_size = []
libs_size = []
for (name, data) in raw_data.items():
    crates = data["cargo_bloat_crates"]["crates"]
    size_tuf_no_std = sum(c["size"] for c in crates if c["name"].startswith("tuf_no_std")) / (1024**1)
    size_ = sum(c["size"] for c in crates if c["name"].startswith("")) / (1024**1)
    size_libs = sum(c["size"] for c in crates if any(c["name"].startswith(l) for l in crypto_libs)) / (1024**1)
    libs_size.append((name, size_libs, size_libs))
    _size.append((name, size_ + size_libs, size_))
    tuf_no_std_size.append((name, size_tuf_no_std + size_ + size_libs, size_tuf_no_std))
wifi_driver_size_data = [(name, data + wifi_driver_size) for (name, data) in file_size_data]
print(file_size_data, wifi_driver_size_data)
tuf_no_std_size = list(zip(*tuf_no_std_size))
_size = list(zip(*_size))
libs_size = list(zip(*libs_size))

table_data = {
    "Entire Binary": [f"{v:.2f} KiB" for (_, v) in file_size_data],
    ".text section": [f"{v:.2f} KiB" for (_, v) in file_text_section],
    "TUF": [f"{v:.2f} KiB" for v in  tuf_no_std_size[2]],
    "BT²X": [f"{v:.2f} KiB" for v in  _size[2]],
    "Libs": [f"{v:.2f} KiB" for v in  libs_size[2]],
}
```

```python
fig, ax = plt.subplots()
x_labels = [
    "audit level 2",
    "audit level 2 (no TUF)",
    "audit level 1 (no TUF)",
    "audit level 0",
]
ax.set_ylabel("Size (KiB)")
ax.bar(*zip(*flash_size_data), label="Total Flash", width=0.85, hatch="*", color=flash_size_data_color, align="edge")
ax.bar(*zip(*wifi_driver_size_data), label="WiFi Drivers", width=0.8, hatch="//", color=wifi_driver_size_data_color, align="edge")
ax.bar(*zip(*file_size_data), label="Entire Binary", width=0.8, color=file_size_data_color, align="edge")
ax.bar(*zip(*file_text_section), label=".text section", width=0.75, hatch="o", color=file_text_section_color, align="edge")
ax.bar(tuf_no_std_size[0], tuf_no_std_size[1], label="TUF", width=0.7, hatch="+", color=tuf_no_std_size_color, align="edge")
ax.bar(_size[0], _size[1], label="BT²X", width=0.7, hatch="//", color=_size_color, align="edge")
ax.bar(libs_size[0], libs_size[1], label="Libs", width=0.7, hatch="\\\\", color=libs_size_color, align="edge")
ax.legend(loc="upper right")

# table = plt.table(
#     colWidths=[0.25, 0.25, 0.25, 0.25],
#     cellText=list(table_data.values()),
#     rowLabels=list(table_data.keys()),
#     colLabels=list(raw_data.keys()),
# )
# table.auto_set_font_size(False)
# table.set_fontsize(8)
plt.subplots_adjust(left=0.2, bottom=0.2)
x_tick_distances = [1*v for v in [0.445, 0.445 + 1*(0.85 + 2*0.075), 0.445 + 2*(0.85 + 2*0.075), 0.445 + 3*(0.85 + 2*0.075)]]
ax.set_xticks(x_tick_distances, labels=x_labels)
fig.subplots_adjust(left=0.18, right=2/1.5, bottom=0.05, top=1.5/1.6, hspace=0.4, wspace=0.3)
fig.savefig("plots/benchmark-size-total-storage.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()

ax.set_ylabel("Size (KiB)")
ax.bar(*zip(*file_size_data), align="edge", label="Entire Binary", width=0.8, color=file_size_data_color)
ax.bar(*zip(*file_text_section), align="edge", label=".text section", width=0.75, hatch="o", color=file_text_section_color)
ax.bar(tuf_no_std_size[0], tuf_no_std_size[1], align="edge", label="TUF", width=0.7, hatch="+", color=tuf_no_std_size_color)
ax.bar(_size[0], _size[1], align="edge", label="BT²X", width=0.7, hatch="//", color=_size_color)
ax.bar(libs_size[0], libs_size[1], align="edge", label="Libs", width=0.7, hatch="\\\\", color=libs_size_color)
ax.legend(loc="upper right")



table = plt.table(
    colWidths=[0.25, 0.25, 0.25, 0.25],
    cellText=list(table_data.values()),
    rowLabels=list(table_data.keys()),
    colLabels=x_labels,
)
table.auto_set_font_size(False)
table.set_fontsize(8)
plt.subplots_adjust(left=0.2, bottom=0.2)
plt.xticks([])
fig.subplots_adjust(left=0.18, right=2/1.6, bottom=0.05, top=1.5/1.6, hspace=0.4, wspace=0.3)
fig.savefig("plots/benchmark-size-used-storage.pdf", bbox_inches='tight')
plt.show()
```

## Notes

- The binaries include no business logic and the rest of the size is mostly from the embassy runtime.

## Analysis

- Using TUF in addition to BT²X introduces some overhead but is able to reuse a lot of the dependencies that are already required by BT²X.
- Overall, the overhead of the verification code is roughly equal to the size of the WiFi drivers. Both together take up roughly 1/4 of the available flash.
- The rest of the storage can be used for business logic and storage such as files and new updates.
